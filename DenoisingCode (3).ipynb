# Image Denoising

Whole code

```python
from PIL import Image, ImageTk, ImageDraw, ImageFilter
import scipy.ndimage as ndimage
import numpy as np
from matplotlib import pyplot as plt
from google.colab import files
import io
import sys

# Create a file uploader widget
uploader = files.upload()

# Retrieve the uploaded file path
file_path = next(iter(uploader))

# Opening Picture
open = Image.open(file_path)
img = Image.open(file_path)

# Displaying Image
plt.imshow(img)
plt.show()

# Inputing values for sigma and areas to be crop
print("Enter the value needed for smoothening the image")
sigma_x = input("Sigma X: ")
sigma_y = input("Sigma Y: ")
sigma_z = input("Sigma Z: ")

if len(sigma_x) == 0 or len(sigma_y) == 0 or len(sigma_z) == 0:
    print("Sigma values inputted are not complete or empty.")
    print("Program will stop")
    sys.exit()

print("enter the are to be crop")
area_t = input("Top Value: ")
area_b = input("Bottom Value: ")
area_l = input("Left Value: ")
area_r  = input("Right Value: ")

if len(area_t) == 0 or len(area_b) == 0 or len(area_l) == 0 or len(area_r) == 0:
    print("Values inputted are not complete ur empty.")
    print("Program will stop")
    sys.exit()

#creating array
img = np.array(img)

#variables
ing_x = float(sigma_x)
ing_y = float(sigma_y)
#ing_z = float(sigma_z)
#ord = int(order)
are_t = int(area_t)
are_b = int(area_b)
are_r = int(area_r)
are_l = int(area_l)

#add noise to image

noise_density = 0.05;
noisy_image = np.copy(img);
h, w , _ = noisy_image.shape;
num_salt_pixels = int(noise_density * h * w);
num_pepper_pixels = int(noise_density * h * w);

# Add salt noise
salt_coords = [np.random.randint(0, h, num_salt_pixels), np.random.randint(0, w, num_salt_pixels)];
noisy_image[salt_coords] = 255;

# Add pepper noise
pepper_coords = [np.random.randint(0, h, num_pepper_pixels), np.random.randint(0, w, num_pepper_pixels)];
noisy_image[pepper_coords] = 0;

# Cropping image
cropped_img = noisy_image[are_t:are_b, are_l:are_r];

#plt.imshow(cropped_img)
#plt.show()

#Applying Gaussian filter on Original Image
    #3d gaussian filter, sigma = smoothness of image,
    #                    order = no of times filter is applied
ing = ndimage.gaussian_filter(noisy_image, sigma=(ing_x, ing_y, 0))

#Cropping the original image
#yak = img.copy()
#smol_img = yak[are_t:are_b, are_l:are_r]
amp = ndimage.gaussian_filter(cropped_img, sigma=(ing_x, ing_y, 0))

#pasting denoised crop image to original image
back_im = open.copy()
back_im.paste(Image.fromarray(amp),(are_l,are_t, are_r,are_b))
back_i = open.copy()
back_i.paste(Image.fromarray(cropped_img),(are_l,are_t, are_r,are_b))

#back_im.save("awit.jpg")
#files.download('awit.jpg')

#output
fig, axs = plt.subplots(3, 2, figsize=(10, 5))
axs[0, 0].imshow(img)
axs[0, 0].set_title('Original Image')
axs[0, 0].axis('off')
axs[1, 0].imshow(ing)
axs[1, 0].set_title('Denoised Image')
axs[1, 0].axis('off')
axs[2, 0].imshow(back_im)
axs[2, 0].set_title('Cropped Denoised Image on Original')
axs[2, 0].axis('off')

axs[0, 1].imshow(noisy_image)
axs[0, 1].set_title('Noisy Image')
axs[0, 1].axis('off')
axs[1, 1].imshow(cropped_img)
axs[1, 1].set_title('Cropped noisy Image')
axs[1, 1].axis('off')
axs[2, 1].imshow(back_i)
axs[2, 1].set_title('Cropped noisy Image on Original')
axs[2, 1].axis('off')

plt.show()
```
